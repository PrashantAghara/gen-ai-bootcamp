```python
import pandas as pd
```

```python
messages = pd.read_csv("/kaggle/input/datasets/prashantaghara/sms-spam/SMSSpamCollection.txt", sep="\t", names=["label", "message"])
```

```python
messages
```

```python
# Data Cleaning & Preprocessing
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

### BAG OF WORDS (BOW)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=100, binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

#### BOW With N-Grams

```python
cv.vocabulary_
```

```python
cv_with_ngram = CountVectorizer(max_features=100, binary=True, ngram_range=(2,3))
```

```python
X = cv_with_ngram.fit_transform(corpus).toarray()
```

```python
cv_with_ngram.vocabulary_
```
