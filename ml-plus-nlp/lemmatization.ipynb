```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
pos = n (Noun), v (Verb), a (Adjective), r (adverb)
'''
lemmatizer.lemmatize("eating")
```

```python
lemmatizer.lemmatize("eating", pos='v')
```

```python
words = ["eating", "eats", "eaten", "writing", "written", "going", "goes", "gone"]
```

```python
for word in words:
    print(f"{word} : {lemmatizer.lemmatize(word, pos='v')}")
```

```python
for word in words:
    print(f"{word} : {lemmatizer.lemmatize(word)}") # default (NOUN)
```

```python
lemmatizer.lemmatize("fairly", pos='r') # Takes more time than Stemming.
```
