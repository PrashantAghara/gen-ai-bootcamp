```python
a = float(input("Enter first Number : "))
b = float(input("Enter second Number : "))
opr = input("Enter the operation from (+, /, -, *): ")

ans = 0
if opr == '+':
    ans = a + b
elif opr == '*':
    ans = a * b
elif opr == '-':
    ans = a - b
elif opr == '/':
    ans = a / b

print("Your answer is : ", ans)
```
