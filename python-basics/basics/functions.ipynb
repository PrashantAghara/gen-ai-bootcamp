```python
def pos_args(*args):
    for arg in args:
        print(arg, end=" ")

pos_args(1, 2, 3, 4)
```

```python
def key_args(**kwargs):
    for key, val in kwargs.items():
        print(key, val)

key_args(name='prashant', age=26)
```

```python
# lambda argument: expression

def addition(a, b):
    return a + b

addition(2,3)
```

```python
addition = lambda a,b: a+b
```

```python
addition(1,3)
```

```python
numbers = [1,2,3,4,5]

list(map(lambda x: x*x, numbers))
```

```python
list(filter(lambda x: x%2==0, numbers))
```
