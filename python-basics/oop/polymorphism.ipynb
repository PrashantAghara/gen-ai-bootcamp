Method Overriding

```python
# Method Overriding
class Animal:
    def speak(self):
        print("Sound of animal")

class Dog(Animal):
    def speak(self):
        print("Dog is barking")

dog = Dog()
dog.speak()
```

```python
class Shape:
    def area(self):
        return "The area of the figure"

class Square(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.height * self.width
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return self.radius * self.radius * 3.14
```

```python
def print_area(shape: Shape):
    print(f"Area is {shape.area()}")

square = Square(2, 2)
circle = Circle(3)

print_area(square)
print_area(circle)
```

Interfaces (Abstract Base classes)

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"

class Bike(Vehicle):
    def start_engine(self):
        return "Bike engine started"
```

```python
def print_engine(vehicle: Vehicle):
    print(f"{vehicle.start_engine()}")

print_engine(Car())
print_engine(Bike())
```
