```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def start(self):
        print("The vehicle started")
    
    @abstractmethod
    def start_engine(self):
        pass
```

```python
class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")
```

```python
def operate_vehicle(vehicle: Vehicle):
    vehicle.start_engine()
```

```python
car = Car()
operate_vehicle(car)
```
