```python
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type    
    def drive(self):
        print(f"The person will drive the {self.engine_type} car")
```

```python
car1 = Car(4, 4, 'Petrol')
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, is_self_drive):
        super().__init__(windows, doors, engine_type)
        self.is_self_drive = is_self_drive
    
    def selfdrive(self):
        print(f"Tesla supports self driving : {self.is_self_drive}")
```

```python
variant1 = Tesla(4, 5, 'electric', False)
variant1.selfdrive()
```

```python
variant1 = Tesla(4, 5, 'electric', True)
variant1.drive()
variant1.selfdrive()
```

Multiple Inheritence

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print("Subclasses must implement this method")
```

```python
class Pet:
    def __init__(self, owner):
        self.owner = owner
```

```python
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
    
    def speak(self):
        print(f"{self.name} is barking")
```

```python
dog = Dog('Moti', "Prashant")
dog.speak()
```
