```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(f"{self.name} is barking")
    
    def get_name(self):
        return self.name

dog1 = Dog('batli', 1)
dog1.bark()
dog1.get_name()
```

```python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, money):
        self.balance += money
        print(f"{money} is deposited, new balance is {self.balance}")
    
    def withdraw(self, money):
        if money > self.balance:
            print("Insufficient Balance")
        else:
            self.balance -= money
            print(f"{money} is withdrawn from the account")
    
    def print_balance(self):
        print(f"Current Balance is {self.balance}")

my_acc = BankAccount('prashant', 1000)
my_acc.print_balance()
my_acc.withdraw(100)
my_acc.print_balance()
my_acc.deposit(100)
my_acc.print_balance()
my_acc.withdraw(1500)
```
