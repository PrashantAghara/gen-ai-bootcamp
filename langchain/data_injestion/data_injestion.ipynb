### Document Loaders

### Text Loaders

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
```

```python
document = loader.load()
document
```

#### Reading from PDF File

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('attention.pdf')
docs = loader.load()
docs
```

```python
type(docs)
```

```python
type(docs[0])
```

#### Web base loaders

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(web_paths =('https://en.wikipedia.org/wiki/Generative_AI',), bs_kwargs=dict(parse_only=bs4.SoupStrainer(
    class_=('firstHeading','vector-body')
)))
```

```python
loader.load()
```

#### Arxiv

```python
from langchain_community.document_loaders import ArxivLoader

loader = ArxivLoader(query="1605.08386", load_max_docs=2)
docs = loader.load()
docs
```

#### Wikipedia Loader

```python
from langchain_community.document_loaders import WikipediaLoader

loader = WikipediaLoader(query='India', load_max_docs=2)
docs = loader.load()
docs
```
