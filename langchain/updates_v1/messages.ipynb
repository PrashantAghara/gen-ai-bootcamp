```python
import os
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = ChatGroq(model="openai/gpt-oss-20b")
```

```python
model.invoke("Hi, how are you?")
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage
```

```python
messages = [
    SystemMessage(content="You are a poetry expert."),
    HumanMessage(content="Write a poem on Artificial Intelligence")
]
```

```python
model.invoke(messages)
```

```python
system_msg = SystemMessage(content="You are a helpful coding assistant")

messages = [
    system_msg,
    HumanMessage(content="How do I create a REST API?")
]

response = model.invoke(messages)

response.content
```

```python
system_msg = SystemMessage(content="""
    You are a Senior Python Developer with expertise in web frameworks. 
    Always provide code examples and explain your reasoning. 
    Be concise but thorough in your explainations.
""")

messages = [
    system_msg,
    HumanMessage(content="How do I create a REST API?")
]

response = model.invoke(messages)
print(response.content)
```

```python
human_msg = HumanMessage(
    content="Hello!",
    name="Prashant",
    id="msg_123"
)
```

```python
model.invoke([
    human_msg
])
```

```python
ai_msg = AIMessage("I'd be happy to help you with that question!")

messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great! What is 2+2")
]

response = model.invoke(messages)

print(response.content)
```
